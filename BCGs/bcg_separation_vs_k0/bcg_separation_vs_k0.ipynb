```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('default')

data = np.genfromtxt(r"../bcg_coords_compare/bcg_coords_compare.csv",delimiter=',',skip_header=1,usecols=[1,4]) #pulls only sep and k0
data = data[~np.isnan(data).any(axis=1)] #drops rows with nans
```

```python
plt.scatter(data[:,0],data[:,1])
plt.xscale('log')
plt.xlabel("Angular Separation of BCG and Cluster Center (arcseconds)")
plt.ylabel("K0 (keV cm2)")
plt.title("Separation of BCG versus Entropy of Cluster")
plt.grid()
plt.savefig('bcg_separation_vs_k0.png')
plt.show()
```
